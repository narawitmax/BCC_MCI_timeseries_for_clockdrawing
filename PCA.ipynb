```python
import sys
import numpy as np
from sklearn.decomposition import PCA
from numpy.random import multivariate_normal
import matplotlib.pyplot as plt
from matplotlib import colors
import seaborn as sns

import pandas as pd
```

```python
data = pd.read_csv("features_filtered.csv", index_col = 0)
labels = pd.read_csv("label.csv", index_col = 0)

display(data)
display(labels)

lb = (np.array(labels))
```

```python
model_PCA = PCA(n_components=2, whiten=False)
cmap = colors.ListedColormap(["blue","red"])

model_PCA.fit(data)

# ดึงเอา principal component (pc) ออกมา
pc1 = model_PCA.components_[0]
pc2 = model_PCA.components_[1]

#print(model_PCA)
#print(pc1)
#print(np.argmax(pc1))
#print(np.max(pc1))
#print(np.argmin(pc1))
#print(np.min(pc1))
#print(pc2)

#plt.bar(np.arange(len(pc1)),pc1)
#plt.xlim(1072,1074)
#plt.show()

#plt.bar(np.arange(len(pc2)),pc2)
#plt.show()

# คำนวณหาความชันของเส้นตรงที่วิ่งผ่านจุด (0,0) และว่ิงไปในทิศทางเดียวกับ principal component -> y = m*x
m_pc1 = pc1[1]/pc1[0]
m_pc2 = pc2[1]/pc2[0]

# คำนวณพิกัดใหม่ของทุกจุดข้อมูล
data_pca = model_PCA.transform(data)
print("PCA shape =", np.shape(data_pca))


print("data_pca =", data_pca)

plt.hist(data_pca[:, 0], 100, density=True, histtype='bar', label=lb)
plt.xlim(-10000000000000000, 50000000000000000)
plt.legend(prop={'size': 10})
plt.show()

# Plot ข้อมูลที่ลดมิติโดยการตัดแกน PC2 ทิ้งไป
plt.figure(figsize=(4, 0.5))
plt.scatter(data_pca[:, 0], np.zeros_like(data_pca[:, 0]), c=lb, cmap = cmap)
#ax[2].plot(x_disp, [0, 0], c="m", alpha=1) # แสดงผลโดยการหมุนแกน pc1 ให้มาอยู่ที่ตำแหน่งแกน x
#plt.xlim(x_disp)
#plt.ylim(-0.5, 0.5)
plt.title("Data in the resulting 1-dimensional space (PC1)")
plt.yticks([])
plt.show()


# Plot ข้อมูลที่ลดมิติโดยการตัดแกน PC1 ทิ้งไป
plt.figure(figsize=(4, 0.5))
plt.scatter(data_pca[:, 1], np.zeros_like(data_pca[:, 1]), c=lb, cmap = cmap)
#ax[2].plot(x_disp, [0, 0], c="m", alpha=0.3) # แสดงผลโดยการหมุนแกน pc2 ให้มาอยู่ที่ตำแหน่งแกน x
#plt.xlim(x_disp)
#plt.ylim(-0.5, 0.5)
plt.title("Data in the resulting 1-dimensional space (PC2)")
plt.yticks([])
plt.show()

print(f"% variance captured by PC1 = {model_PCA.explained_variance_ratio_[0]}")
print(f"% variance captured by PC2 = {model_PCA.explained_variance_ratio_[1]}")
```
